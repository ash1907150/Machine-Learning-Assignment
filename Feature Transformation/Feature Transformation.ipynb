# Exploring Scaling For feature Transformation

Importing Necessary Libary

```python
import numpy as np
import pandas as pd
```

Reading The CSV File

```python
df = pd.read_csv('supershops.csv')
```

Check for NaNs

```python
df.isnull().sum()
```

Dropping the Row Where Data is Missing

```python
df.dropna(inplace = True )
df.isnull().sum()
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df_norm = df.copy()
model_norm = MinMaxScaler()
```

```python
df_norm['Marketing Spend'] = model_norm.fit_transform(df_norm[['Marketing Spend']]) # fit() = training the model ; transform() = replace the values
df_norm['Administration'] = model_norm.fit_transform(df_norm[['Administration']])
df_norm['Transport'] = model_norm.fit_transform(df_norm[['Transport']])
df_norm['Profit'] = model_norm.fit_transform(df_norm[['Profit']])
```

```python
df_norm.head()
```

```python
df_norm.dropna()
```
